#Preparation

```python
from scipy.stats import zscore

# Menormalisasi kolom 'Sales' dan 'Profit' menggunakan Z-scores
data['Sales_Zscore'] = zscore(data['Sales'])
data['Profit_Zscore'] = zscore(data['Profit'])
data
```

```python
# Diskretisasi kuantitas
def categorize_quantity(row):
    categories = []
    for item, qty in zip(row['Product Name'], row['Quantity']):
        if qty <= 2:
            categories.append(f"{item} - Low")
        elif 3 <= qty <= 5:
            categories.append(f"{item} - Medium")
        else:
            categories.append(f"{item} - High")
    return categories

# Terapkan diskretisasi
transaction_data_qty = data.groupby('Order ID').apply(categorize_quantity)
```

```python
transaction_data_qty
```

```python
te = TransactionEncoder()
te_array = te.fit(transaction_data_qty).transform(transaction_data_qty)
df_qty = pd.DataFrame(te_array, columns=te.columns_)
```

```python
df_qty
```

#Profit

```python
# Tambahkan kolom kategori Z-score untuk keperluan analisis
def categorize_zscore(z):
    if z < -1:
        return "Low"
    elif -1 <= z <= 1:
        return "Medium"
    else:
        return "High"

data['Profit Z-score Category'] = data['Profit_Zscore'].apply(categorize_zscore)
```

```python
# Gabungkan nama produk dengan kategori Z-score
data['Product With Z-score'] = data['Product Name'] + " - " + data['Profit Z-score Category']

# Kelompokkan data berdasarkan Order ID
transaction_data = data.groupby('Order ID')['Product With Z-score'].apply(list)

# Transformasi data transaksi ke dalam bentuk untuk fpgrowth
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transaction_data).transform(transaction_data)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Jalankan algoritma fpgrowth
from mlxtend.frequent_patterns import fpgrowth
frequent_itemsets_fp = fpgrowth(df, min_support=0.0006, use_colnames=True)

# Output frequent itemsets
print(frequent_itemsets_fp.head())
```

#Memperhatikan Profit

```python
# Hitung rata-rata Z-score untuk setiap produk
product_zscore = data.groupby('Product Name')['Profit_Zscore'].mean().to_dict()

# Tambahkan rata-rata Z-score ke itemsets
def calculate_itemset_zscore(itemset):
    return sum([product_zscore.get(item.split(' - ')[0], 0) for item in itemset]) / len(itemset)

frequent_itemsets_fp['Average Z-score'] = frequent_itemsets_fp['itemsets'].apply(calculate_itemset_zscore)

# Output frequent itemsets dengan Z-score
frequent_itemsets_fp
```

```python
from mlxtend.frequent_patterns import association_rules

# Buat association rules
rules = association_rules(frequent_itemsets_fp, num_itemsets=len(frequent_itemsets_fp), metric="lift", min_threshold=1.0)

# Tambahkan rata-rata Z-score untuk antecedents dan consequents
def calculate_rule_zscore(items):
    return sum([product_zscore.get(item.split(' - ')[0], 0) for item in items]) / len(items)

rules['antecedents_avg_zscore'] = rules['antecedents'].apply(lambda x: calculate_rule_zscore(list(x)))
rules['consequents_avg_zscore'] = rules['consequents'].apply(lambda x: calculate_rule_zscore(list(x)))

# Output rules dengan Z-score
rules
```
